```python
import keras
from keras.models import Sequential
from keras.layers import Dense,Dropout,Activation
from keras.optimizers import SGD
import numpy as np
import pandas as pd
```

```python
def pdf(x,tar):
    mean = np.mean(x)
    std = np.std(x)+tar
    y_out = 1/(std * np.sqrt(2 * np.pi)) * np.exp( - (x - mean)**2 / (2 * std**2))
    return y_out
```

```python
a=2
b=10

#200 x's
m2 = 0.04

x = np.arange(a, b, m2)

print("L--------")
#Low concentration
low = 0.10
inc = 0

k=0
for i in range(20):
    #print(i)
    y = pdf(x,low+inc)
    y = np.insert(y, 0, 0, axis=0)
    
    if(i==0):
        data = {"amplitude 0":y}
        df = pd.DataFrame(data)
    else:    
        name = "amplitude "+str(k)
        df[name] = y
    
    
    inc = inc+0.005
    k+=1
    
print("M--------")
#Medium concentration
medium = 0.40
inc = 0
for i in range(20):
    #print(medium+inc)
    y = pdf(x,medium+inc)
    y = np.insert(y, 0, 1, axis=0)
    
    name = "amplitude "+str(k)
    df[name] = y
    
    inc = inc+0.005
    

    k+=1
    
    
print("H--------")
#High concentration
high = 0.60
inc = 0.0 
for i in range(20):
    #print(high+inc)
    y = pdf(x, high+inc)
    y = np.insert(y, 0, 2, axis=0)
    
    name = "amplitude "+str(k)
    df[name] = y
    
    inc = inc+0.005
    k+=1
    
#print(y)
df 
```

```python
df_df = pd.DataFrame(np.array(df).T)
df_df.head(10)
```

```python
df_shuffle = df_df.sample(frac=1).reset_index(drop=True)
```

#### Set: Categorical Y train and testm

```python
Y_train = keras.utils.to_categorical(Y_train, num_classes=3,dtype="int32")
y_test = keras.utils.to_categorical(y_test, num_classes=3,dtype="int32")
```

```python
#Model
```

```python
model = Sequential()
```

```python
model.add(Dense(64, activation='relu', input_dim=200))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(3, activation='softmax'))

sgd = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
```

```python
model.compile(loss='categorical_crossentropy',optimizer=sgd,metrics=['accuracy'])
```

```python
model.fit(X_train,Y_train,epochs=2000,batch_size=128)
```

```python
score = model.evaluate(x_test, y_test, batch_size=128,)
```
