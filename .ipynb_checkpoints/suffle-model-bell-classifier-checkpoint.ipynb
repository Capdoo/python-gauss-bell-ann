```python
import keras
from keras.models import Sequential
from keras.layers import Dense,Dropout,Activation
from keras.optimizers import SGD
import numpy as np
import pandas as pd
```

```python
def pdf(x,tar):
    mean = np.mean(x)
    std = np.std(x)+tar
    y_out = 1/(std * np.sqrt(2 * np.pi)) * np.exp( - (x - mean)**2 / (2 * std**2))
    return y_out
```

```python
a=2
b=10

#200 x's
m2 = 0.04

x = np.arange(a, b, m2)

print("L--------")
#Low concentration
low = 0.10
inc = 0

k=0
for i in range(20):
    #print(i)
    y = pdf(x,low+inc)
    y = np.insert(y, 0, 0, axis=0)
    
    if(i==0):
        data = {"amplitude 0":y}
        df = pd.DataFrame(data)
    else:    
        name = "amplitude "+str(k)
        df[name] = y
    
    
    inc = inc+0.005
    k+=1
    
print("M--------")
#Medium concentration
medium = 0.40
inc = 0
for i in range(20):
    #print(medium+inc)
    y = pdf(x,medium+inc)
    y = np.insert(y, 0, 1, axis=0)
    
    name = "amplitude "+str(k)
    df[name] = y
    
    inc = inc+0.005
    

    k+=1
    
    
print("H--------")
#High concentration
high = 0.60
inc = 0.0 
for i in range(20):
    #print(high+inc)
    y = pdf(x, high+inc)
    y = np.insert(y, 0, 2, axis=0)
    
    name = "amplitude "+str(k)
    df[name] = y
    
    inc = inc+0.005
    k+=1
    
#print(y)
df 
```
